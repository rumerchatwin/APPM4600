```python
# question 4
w = 10**(-np.linspace(1,10,10))
```

```python
w
```

```python
# w is an array of values of 10 to the power of a negative number from 1-10
X = np.arange(1,11,1)
plt.semilogy(X, w)
plt.xlabel('X')
plt.ylabel('w')
plt.title('The Plot of X vs w')
```

```python
S = 3*w
plt.semilogy(X,S)
plt.semilogy(X, w)
plt.xlabel('X')
plt.ylabel('S')
plt.title('The Plot of X vs S')
```
